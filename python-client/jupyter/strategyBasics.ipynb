```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from binance.client import Client
from history.retrieve import RetrieveHistory
from services.setup import ClientSetup
from services.rest import BinanceRest
from utils.constants import AppConstants
from utils.helpers.convert import get_datetime
from utils.constants import DateTimeConstants
```

```python
symbol = "XRPUSDT"
interval=Client.KLINE_INTERVAL_1HOUR
```

```python
client = ClientSetup(AppConstants.ENV_TEST).setup()
rest = BinanceRest(symbol, client)
```

```python
history = RetrieveHistory(symbol, interval, client)
```

```python
data = history.get_historical_klines("1 day ago UTC")
```

```python
df = pd.DataFrame(data)
df
```

```python
# set column names
col = columns=['Open Time','Open','High','Low','Close','Volume','Close Time','Quote Asset Volume','Number Of Trades','Taker buy base asset volume','Taker buy quote asset volume','ignore']
df.columns=col
df
```

```python
df.dtypes
```

```python
# get mean of close prices
df['Close'].astype(float).mean()
```

```python
print(get_datetime(1595116800000,DateTimeConstants.MS))
```

```python
# use our converter to display user friendly date values
rowDateIndex = df['Open Time'].apply(lambda x: get_datetime(x,DateTimeConstants.MS)) 
rowDateIndex
```

```python
# assign the datetime index as a row index for dataFrame
df.index = rowDateIndex
df
```

```python
# get rid of columns in dataFrame that we dont need
df = df.drop(columns=['Open Time', 'Close Time','Quote Asset Volume','Taker buy base asset volume','Taker buy quote asset volume','ignore'])
df
```

```python
# set some types on the columns we want to keep
df['Open'] = df['Open'].astype(float)
df['High'] = df['High'].astype(float)
df['Low'] = df['Low'].astype(float)
df['Close'] = df['Close'].astype(float)
df['Volume'] = df['Volume'].astype(float)
df['Number Of Trades'] = df['Number Of Trades'].astype(int)
# df.dtypes
df
```

```python
# suppress scientific notations ex: e+07 
pd.options.display.float_format = '{:.9}'.format 
df.describe()
# reset suppress:
# pd.reset_option('^display.', silent=True)
```
