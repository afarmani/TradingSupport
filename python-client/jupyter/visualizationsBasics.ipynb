```python
from pylab import plt, mpl
import numpy as np
import pandas as pd
```

```python
# set plotting styles
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
# create random dataset to play with
np.random.seed
randvalues = np.random.standard_normal((9, 4))
dates = pd.date_range('2020-1-1', periods=9,freq='M')
df = pd.DataFrame(randvalues, )
df.columns = ['No1','No2','No3','No4']
df.index = dates
df
```

```python
# Lineplot
df.cumsum().plot(lw=2.0,figsize=(10,6))
```

```python
# bar plot
df.cumsum().plot.bar(figsize=(10,6),rot=15)
# alternative 
```

```python
# documentation https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html
df.plot(kind='box',figsize=(10,6))
```
