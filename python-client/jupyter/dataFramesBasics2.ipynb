```python
# examples based on ndarray object & features DatetimeIndex to manage time series data
import numpy as np
import pandas as pd
```

```python
np.random.seed
randvalues = np.random.standard_normal((9, 5)) # creates 5 entries of 9 decimals
randvalues
```

```python
# takes ndarray object and sets "meta-info (index info)"
df = pd.DataFrame(randvalues)
df
```

```python
df.dtypes
```

```python
df.columns=['no1','no2','no3','no4','no5']
df
```

```python
# Generate a DatetimeIndex Object (you can pass start and end time as well)
dates = pd.date_range('2019-1-1', periods=9,freq='M')
dates
```

```python
# take the DatetimeIndex object created and assign it as the index for the dataFrame
df.index = dates
df
```

```python
# generates a dateFrame object from ndarray
df.values
```

```python
# generate a ndarray from dataFrame
np.array(df)
```

```python
# retrieve meta-info regarding your obect, using built in methods
df.info()
```

```python
# provides statistics for each column
df.describe()
```

```python
# column wise sums
df.sum()
```

```python
# row size sums
df.sum(axis=1)
```

```python
# cumulative sum of columns
df.cumsum()
```

```python
# we can also use NumPy built in methods on dataFrames
np.mean(df)
```

```python
# GROUPBY OPERATIONS
df['Quarters'] = ['Q1','Q1','Q1','Q2','Q2','Q2','Q3','Q3','Q3']
df
```

```python
groups = df.groupby('Quarters')
# groups.size()
# groups.mean()
groups.aggregate([min,max]).round(2) # Gets min max per column & round decimal points 2
```

```python
df['Odd_Even'] = ['Even','Odd','Even','Odd','Even','Odd','Even','Odd','Even']
df
```

```python
# multi column grouping
groups = df.groupby(['Quarters', 'Odd_Even'])
groups.size()
```

```python
groups[['no1','no2']].aggregate([min, max])
# groups.aggregate([sum, np.mean])
```
